For further analysis, we need the max order of users, the reorder ratio of users and average delay in consecutive order for a user in the prior dataframe:

```python
max_orders_user=df_prior_new.groupby('user_id')['order_number'].max().reset_index()
```

```python
reorder_ratio_users=df_prior_new.groupby('user_id')['reordered'].mean().reset_index()
```

```python
reorder_ratio_users.head()
```

```python
days_prior_user=df_prior_new.groupby('user_id')['days_since_prior_order'].mean().reset_index()
days_prior_user.head()
```

```python
#ordering by product id
max_orders_prod=df_prior_new.groupby('product_id')['order_id'].count().reset_index()
max_orders_prod.head()
```

```python
#ordering by product id
reorder_ratio_prod=df_prior_new.groupby('product_id')['reordered'].mean().reset_index()
reorder_ratio_prod.head()
```

```python
#ordering by product id
days_prior_prod=df_prior_new.groupby('product_id')['days_since_prior_order'].mean().reset_index()
days_prior_prod.head()
```

<h1>Model:</h1>

```python
#creating a classification matrix for xgboost
data_dmatrix=xgboost.DMatrix(data=X_train,label=y_train)
```

```python
data_dmatrix
```

```python
#hyperparameter tuning
xgb_params = {
    "objective"         : "reg:logistic"
    ,"eval_metric"      : "logloss"
    ,"eta"              : 0.2
    ,"max_depth"        : 7
    ,"min_child_weight" :10
    ,"gamma"            :0.5
    ,"subsample"        :0.7
    ,"colsample_bytree" :0.95
    ,"alpha"            :1e-05
    ,"lambda"           :8
}
#training and plotting the loss
watchlist= [(data_dmatrix, "train")]
xg_reg = xgboost.train(params=xgb_params, dtrain=data_dmatrix, num_boost_round=80, evals=watchlist, verbose_eval=10)
xgboost.plot_importance(xg_reg)
```

```python
data_test.drop(columns=['order_id','order_id_y','eval_set'],inplace=True)
data_test.set_index(['user_id','product_id'])
```

```python
pred=xgboost.DMatrix(data_test[['reorder_ratio','days_since_prior_order','order_number']])
```

```python
#getting predictions on the test data in the correct format
new_pred=xg_reg.predict(pred)
```

```python
#classifying users on the basis of a threshold
data_test['predictions']=((new_pred) > 0.2).astype(int)
```

```python
orders_test = df_orders[df_orders.eval_set=='test']
prediction = data_test.merge(orders_test[["user_id", "order_id"]], on='user_id', how='left')
prediction.head()
```

```python
#grouping the orders and the products together if they were reordered
d = dict()
for row in prediction.itertuples():
    if row.predictions == 1:
        try:
            d[row.order_id] += ' ' + str(row.product_id)
        except:
            d[row.order_id] = str(row.product_id)

for order in prediction.order_id:
    if order not in d:
        d[order] = np.NaN
        
gc.collect()
```

```python
#resultant df
fin_df = pd.DataFrame.from_dict(d, orient='index')
fin_df.reset_index(inplace=True)
fin_df.columns = ['order_id', 'products']

fin_df.head()
```
